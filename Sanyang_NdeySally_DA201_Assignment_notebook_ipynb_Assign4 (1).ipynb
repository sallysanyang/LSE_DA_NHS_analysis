```python
import seaborn as sb
import matplotlib as plt
import matplotlib.pyplot as py_plt
import pandas as pd
```

```python
pd_csv=pd.read_csv('national_categories.csv')
sb.set(rc={'figure.figsize':(30,25)},style='white')
```

```python
# referencing the columns in the .csv file
pd_csv
```

```python
#change data type of appointment month to string
pd_csv['appointment_month'].astype(str)
```

```python
#Aggregate the appointments per month and determine the sum of the appointments per month
nc_ss=pd.read_csv('national_categories.csv')
group_by=nc_ss.groupby(['appointment_month','service_setting']).sum()
new_df=group_by.reset_index()
new_df
```

```python
# visualisation of types of appointment and their count
s_plt=sb.lineplot(x=new_df['appointment_month'],y=new_df['count_of_appointments'],hue=new_df['service_setting'],data=nc_ss)
```

```python
# Repeat Objective 1 (b) and (c) for context types and national categories columns

# (a) context types 
nc_ct=pd.read_csv('national_categories.csv')
g_by=nc_ct.groupby(['appointment_month','context_type']).sum()
new_df=g_by.reset_index()
new_df
```

```python
s_plt=sb.lineplot(x=new_df['appointment_month'],y=new_df['count_of_appointments'],hue=new_df['context_type'],data=nc_ct)
```

```python
# (a) national categories
nc_nc=pd.read_csv("national_categories.csv")
g_by=nc_nc.groupby(['appointment_month','national_category']).sum()
new_df=g_by.reset_index()
new_df
```

```python
s_plt=sb.lineplot(x=new_df['appointment_month'],y=new_df['count_of_appointments'],hue=new_df['national_category'],data=nc_nc, errorbar=None)
```

```python
nc_ss_day=pd.read_csv('national_categories.csv')
g_by=nc_ss_day.groupby(['appointment_date','appointment_month','service_setting']).sum()
new_df=g_by.reset_index()
new_df['winter']=new_df['appointment_month'].apply(lambda x:x if x.split('-')[-1]=='01' else False)
winter_df=new_df[new_df['winter']!=False]
winter_df
```

```python
#visualisations indicating the number of appointments for service setting in winter
s_plt=sb.lineplot(x=winter_df['appointment_date'],y=winter_df['count_of_appointments'],hue=winter_df['service_setting'],data=nc_ss_day, errorbar=None).set(title="2022-01")
py_plt.xticks(rotation=45)
```
